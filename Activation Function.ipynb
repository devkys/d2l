# Step function

```python
import numpy as np
x = np.array([-1.0, 1.0, 2.0])
y = x > 0
print(y)   # ----------> 원소 각각에 부등호 연산을 수행하면 bool배열이 생성
y = y.astype(np.int32)  # ------------> bool타입을 넘파이 astype을 이용해서 정수형으로 변환
print(y)
```

### 계단함수 단순하게 구현
#### 이 구현은 단순하고 쉽지만, 인수 x는 실수(부동 소수점)만 받아들임
#### <font color=red>즉, 넘파이 배열을 인수로 받아들이지 못함</font>

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

### step function 정의

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x > 0, dtype=np.int32)
```

### 계단 함수는 0을 경계로 출력이 0에서 1(또는 1에서 0)

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.title('step function')
plt.ylim(-0.1, 1.1)
plt.show()
```

# Sigmoid function

```python
# sigmoid function 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### sigmoid 함수는 입력에 따라 출력이 연속적으로 변화

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.title("sigmoid")
plt.ylim(-0.1, 1.1)
plt.show()
```

## Step function vs Sigmoid function

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x)
y2 = sigmoid(x)
plt.plot(x,y1, label="step", linestyle="--")
plt.plot(x,y2, label="sigmoid")
plt.xlabel("x")
plt.ylabel("y")
plt.title("step & sigmoid function")
plt.legend()
plt.show
```

#### 둘 함수의 가장 큰 차이는 형태 즉, 매끄러움
#### 계단 함수는 0을 경계로 출력이 갑자기 바뀜
#### sigmoid 함수의 매끈함은 신경망 학습에서 아주 중요한 역할을 함
#### 큰 관점에서 보면 둘은 같은 모양을 하고 있음
#### 둘 다 입력이 작을 때의 출력은 0에 가깝고(혹은 0), 입력이 커지면 출력이 1에 가까워지는(혹은 1)
### 즉, 계단 함수와 시그모이드 함수는 입력이 중요하면 큰 값을 출력하고 입력이 중요하지 않으면 작은 값을 출력

# ReLU function
## 정류된 선형 함수
## 입력이 0을 넘으면 그 입력을 그대로 출력하고, 0이하면 0을 출력하는 함수

```python
# ReLu function 정의
def relu(x):
    return np.maximum(0, x)   # ---------->maximum함수는 두 입력 중 큰 값을 선택해 반환하는 함수

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title("ReLu")
plt.ylim(-1.0, 6.0)
plt.show()
```
